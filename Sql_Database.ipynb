### Connecting to the database named as "fananalytics"

```python
import time
import pandas as pd
```

### Created a table in my database named "matches" to hold all the information corresponding to match summary of all IPL matches

The matches.csv file was imported to the database using the piece of code below and all the columns of the table are also specified.

```python



cursor=db.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS `fananalytics`.`matches` (`0` varchar(3), `Unnamed: 0` varchar(10), `winner_wickets` varchar(14), `umpire3` varchar(7), `team` varchar(27), `team2` varchar(27), `gender` varchar(6), `season` varchar(7), `date` varchar(10), `competition` varchar(21), `match_number` varchar(12), `venue` varchar(52), `city` varchar(14), `toss_winner` varchar(27), `toss_decision` varchar(13), `player_of_match` varchar(17), `umpire` varchar(21), `umpire2` varchar(15), `reserve_umpire` varchar(21), `tv_umpire` varchar(21), `match_referee` varchar(17), `winner` varchar(27), `winner_runs` varchar(11), `outcome` varchar(9), `eliminator` varchar(27), `date2` varchar(10), `neutralvenue` varchar(12), `id` varchar(3)) DEFAULT CHARACTER SET utf8 COLLATE utf8_general_ci;")


# sql = "LOAD DATA LOCAL INFILE '{}' INTO TABLE innings FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"' LINES TERMINATED BY '\n' IGNORE 1 LINES;;""
load_sql = """LOAD DATA INFILE '{}'
         INTO TABLE matches
         FIELDS TERMINATED BY ','
         OPTIONALLY ENCLOSED BY '"'
         ESCAPED BY '"'
         LINES TERMINATED BY '\n'
         IGNORE 1 LINES;"""
cursor = db.cursor()
#             time.sleep(5)
cursor.execute(load_sql.format('matches.csv'))
db.commit()
```

### Created a table in my database named "innings" to hold information of all deliveries of the IPL matches

The innings.csv file was imported to the database using the piece of code below and all the columns of the table are also specified. 

```python
cursor=db.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS `fananalytics`.`innings` (`0` varchar(3), `inning` varchar(50), `over` varchar(50), `batting_team_name` varchar(50), `batsman` varchar(50), `non_striker` varchar(50), `bowler` varchar(50), `runs_off_bat` varchar(50), `Extras` varchar(50), `Kind of wicket` varchar(50), `Dismissed player` varchar(50), `match_id` varchar(10)) DEFAULT CHARACTER SET utf8 COLLATE utf8_general_ci;")


# sql = "LOAD DATA LOCAL INFILE '{}' INTO TABLE innings FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"' LINES TERMINATED BY '\n' IGNORE 1 LINES;;""
load_sql = """LOAD DATA INFILE '{}'
         INTO TABLE innings
         FIELDS TERMINATED BY ','
         OPTIONALLY ENCLOSED BY '"'
         ESCAPED BY '"'
         LINES TERMINATED BY '\n'
         IGNORE 1 LINES;"""
cursor = db.cursor()
#             time.sleep(5)
cursor.execute(load_sql.format('innings.csv'))
db.commit()
```

```python
query = 'select * from matches'
matches = pd.read_sql_query(query, db)
matches.to_csv("matches.csv")
```
